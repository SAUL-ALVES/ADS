```python
#Enumerate
xyz = ['x','y','z']

for i, value in enumerate(xyz):
  print(f"{i}: {value}")
```

```python
#Zip
list1 = ['a','b','c','d','e','f','g']
list2 = [10, 20, 30, 40, 50]
list3 = ['#123', '#456', '#789']

for par in zip(list1, list2, list3):
  print(par)
```

```python
#Range
numeros = [45, 34, 1, 2, 3]
numeros_2 = numeros[3:5]
print(numeros_2)
```

```python
#Range

print(list(range(4)))

print(list(range(1,5)))

print(list(range(1, 10, 2)))

print(list(range(-5)))

print(list(range(10, 5)))

print(list(range(1, 10, -2)))
```

```python
# Exercícios com enumerate

# 1
chico_altura = 1.50
ze_altura = 1.10
chico_crescimento = 0.02
ze_crescimento = 0.03

anos = range(1, 100)

for ano, (chico, ze) in enumerate(zip(
    [chico_altura + chico_crescimento * i for i in anos],
    [ze_altura + ze_crescimento * i for i in anos]), 1):
    if ze > chico:
       print(f'Após {ano} ano(s), Zé será maior que Chico!')
        break
# 2

transacoes = [49, 40, -32, 11, 18, 9, -32, 1, -8, -34,
-39, -30, -33, 38, 36, 35, -46, 16, 43, -23,
44, -22, 50, -50, 46, 24, 49, -42, 12, -1]

for dia, valor in enumerate(transacoes, start = 1):
  print (f'No dia {dia} transação de R$ {valor}')

```
